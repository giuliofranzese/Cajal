## 1. Simulating a Gene Regulatory Network

We simulate a small GRN and then recover its structure using DD-mec.

```python
def simulate_grn(W, T=500, dt=0.05, noise=0.1):
    """Simulate gene regulatory network: dx/dt = W * sigmoid(x) - x + noise"""
    n = W.shape[0]
    traj = np.zeros((T, n))
    traj[0] = np.random.randn(n)
    for t in range(1, T):
        x = traj[t-1]
        dx = W @ (1/(1+np.exp(-x))) - x
        traj[t] = x + dt*dx + noise*np.sqrt(dt)*np.random.randn(n)
    return traj

# Simple 3-gene network: gene 0 -> gene 1 -> gene 2
W_true = np.array([
    [ 0.0,  0.0,  0.0],
    [ 1.5,  0.0,  0.0],
    [ 0.0,  1.5,  0.0],
])

traj = simulate_grn(W_true, T=2000)

fig, axes = plt.subplots(3, 1, figsize=(12, 5), sharex=True)
for i, ax in enumerate(axes):
    ax.plot(traj[:300, i], label=f'Gene {i}')
    ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
