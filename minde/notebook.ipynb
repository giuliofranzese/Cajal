## 1. Correlated Gaussian Benchmark

For bivariate Gaussians, the MI is known analytically: $I(X;Y) = -\frac{1}{2}\log(1-\rho^2)$

```python
def sample_correlated_gaussian(rho, n=1000):
    cov = [[1, rho], [rho, 1]]
    samples = np.random.multivariate_normal([0, 0], cov, n)
    return torch.FloatTensor(samples[:, :1]), torch.FloatTensor(samples[:, 1:])

def gaussian_mi(rho):
    return -0.5 * np.log(1 - rho**2)

rhos = np.linspace(0.1, 0.9, 9)
true_mi = [gaussian_mi(r) for r in rhos]
print("True MI values:", [f"{mi:.3f}" for mi in true_mi])
```
