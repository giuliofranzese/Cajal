## 1. Analytical KL for Gaussians

We verify our estimator against the closed-form solution for Gaussian distributions.

```python
def gaussian_kl(mu1, sigma1, mu2, sigma2):
    return np.log(sigma2/sigma1) + (sigma1**2 + (mu1-mu2)**2)/(2*sigma2**2) - 0.5

# Test: KL between N(0,1) and N(mu, 1) for varying mu
mus = np.linspace(0, 3, 10)
true_kls = [gaussian_kl(0, 1, mu, 1) for mu in mus]
print("True KL values:", [f"{kl:.3f}" for kl in true_kls])
```

## 2. Donsker-Varadhan Estimator

```python
class DVEstimator(nn.Module):
    def __init__(self, dim=1, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim, hidden), nn.ReLU(),
            nn.Linear(hidden, hidden), nn.ReLU(),
            nn.Linear(hidden, 1)
        )
    
    def estimate(self, x_p, x_q):
        t_p = self.net(x_p).squeeze(-1)
        t_q = self.net(x_q).squeeze(-1)
        return t_p.mean() - torch.log(torch.exp(t_q).mean())
    
    def loss(self, x_p, x_q):
        return -self.estimate(x_p, x_q)

# TODO: train and compare to ground truth
pass
```
