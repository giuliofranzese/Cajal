```python
# ---------------------------------------------------------
# Concept Visualization: Trial Structure & Filtering
# ---------------------------------------------------------
# We visualize how we distinguish 'Change' vs 'Non-Change'
flashes = np.arange(1, 12)
image_sequence = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B']
is_change = [False, False, False, False, False, False, False, True, False, False, False]
valid_non_change = [False, False, False, True, True, True, True, False, False, False, False] # Flashes 4 to 7 before change

fig, ax = plt.subplots(figsize=(12, 3))
ax.scatter(flashes[valid_non_change], [1]*sum(valid_non_change), color='blue', s=200, label='Valid Non-Change (Flash 4-10)')
ax.scatter(flashes[np.array(is_change)], [1]*sum(is_change), color='red', marker='*', s=300, label='Valid Change')
ax.scatter(flashes[~np.array(valid_non_change) & ~np.array(is_change)], [1]*sum(~np.array(valid_non_change) & ~np.array(is_change)), color='grey', alpha=0.3, s=100, label='Ignored Flashes')

for i, txt in enumerate(image_sequence):
    ax.annotate(txt, (flashes[i], 1.05), fontsize=14, ha='center')

ax.set_yticks([])
ax.set_xticks(flashes)
ax.set_xlabel("Flash Sequence Number")
ax.set_title("Trial Filtering: Extracting clean stimulus epochs")
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
sns.despine(left=True)
plt.show()
```

```python
def makebins_demo(spike_times, start_time, nb_bins=5, windowDur=50, unit=0.001):
    """
    Demonstration of the binning logic used in download.py.
    Calculates spike counts in consecutive time bins.
    """
    bins = np.zeros((nb_bins))
    bin_size_ms = windowDur * unit # e.g., 50 * 0.001 = 0.05 seconds
    
    for b in range(nb_bins):
        start_ind = np.searchsorted(spike_times, start_time + b * bin_size_ms)
        end_ind = np.searchsorted(spike_times, start_time + (b + 1) * bin_size_ms)
        bins[b] = end_ind - start_ind
    return bins

# Simulate spikes for a single unit around a flash (t=10.0s)
simulated_spikes = np.sort(np.random.uniform(9.9, 10.5, 100))
binned_spikes = makebins_demo(simulated_spikes, start_time=10.0)

plt.figure(figsize=(8, 4))
plt.bar(np.arange(len(binned_spikes)), binned_spikes, color='teal', alpha=0.7)
plt.title("Binned Spike Counts for a Single Unit (Window: 250ms)")
plt.xlabel("Bin Index (50ms per bin)")
plt.ylabel("Spike Count")
plt.show()
```
