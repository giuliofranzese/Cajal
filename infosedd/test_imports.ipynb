```python
import importlib, sys
modules = ['numpy','torch','pytorch_lightning','hydra','omegaconf','Bio','causal_conv1d','mamba_ssm','flash_attn', 'infosedd_synthetic', 'infosedd_real_data']
failed = False
for m in modules:
    try:
        mod = importlib.import_module(m)
        ver = getattr(mod, '__version__', None) or 'unknown'
        print(f"{m}: OK (version: {ver})")
    except Exception as e:
        print(f"{m}: FAILED ({type(e).__name__}: {e})")
        failed = True
if failed:
    sys.exit(2)
```
